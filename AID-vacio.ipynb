### 3.2 Medidas de posición / tendencia central
- Mediana _median()_
- Media aritmética _mean()_
- Moda _mode()_
- Percentiles _quantile()_

```python
print ("Mediana\t\t", df.Precio.median())
print ("Media aritmética ",df.Precio.mean()) 
print()
print ("Moda (Precio)\t\t", df.Precio.mode())
print ("¿Cuántos 5572 hay en Precio", len(df[df.Precio == 5572]))
print ("¿Cuántos 6229 hay en Precio", len(df[df.Precio == 6229]))
print ()
print ("Moda (Traccion)\t\t", df.Traccion.mode())
print ("¿Cuántos 'fwd' hay en Traccion", len(df[df.Traccion == "fwd"]))
print ()
print ("Cuantiles")
print ("Percentil 50",df.Precio.quantile()) #quantile() sin parámetoas se ejecuta con 0.5
print ("Percentil 10",df.Precio.quantile(q=0.1)) #q=0.1 es el primer decil (percentil 10)
print ("Percentil 75",df.Precio.quantile(q=0.75)) #q=0.75 es el tercer cuartil (percentil 75)

```

### 3.3 Medidas de dispersión
- Mínimo _min()_
- Máximo _max()_
- Varianza _var()_
- Desviación estándar _std()_

```python
print ("Mínimo\t", df.Precio.min())
print ("Máximo\t", df.Precio.max())
print()
print ("Varianza\t\t", df.Precio.var())
print ("Desviación estándar\t", df.Precio.std())  #n-1, si se quiere n entonces ddof=0 (grados de libertad)
print()
```

#### 3.3.1 Coeficiente de variación o coeficiente de variación de Pearson
Variación con respecto a la media (si se multiplica por 100 se obtiene un porcentaje) y representa qué tan dispersos están los datos. Entre dos conjuntos de datos, el que tenga un número mayor significa que es más disperso.
$$\dfrac{\sigma}{\bar{x}} $$

```python
print ("Coeficiente de variación")
print ("Largo\t", 100*df.Largo.std() / df.Largo.mean(), "%")
print ("Ancho\t", 100*df.Ancho.std() / df.Ancho.mean(), "%") #Menos disperso
print ("Precio\t", 100*df.Precio.std() / df.Precio.mean(), "%")  #Más disperso
print ("Puertas\t", 100*df.Puertas.std() / df.Puertas.mean(), "%")
```

### 3.4 Medidas de forma
- Asimetría _skew()_
- Curtosis _kurtosis()_ y _kurt()_

```python
print ("Asimetría\t",df.Precio.skew())
print ("Kurtosis\t",df.Precio.kurtosis(),"\t",df.Precio.kurt())
```
